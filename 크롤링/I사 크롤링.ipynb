```python
from selenium import webdriver 
import requests
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Edge()

icecream_url = "주소를 입력하세요"

driver.get(icecream_url)
response = requests.get(icecream_url)
html_content = response.text
soup = BeautifulSoup(html_content, "html.parser")
```

```python
# 각 분류에 해당되는 주소

#학습지도
sort1 = '//*[@id="content"]/div[2]/ul/li[2]/a/span'
#교과지도
sort2 = '//*[@id="content"]/div[2]/ul/li[3]/a/span'
#생활지도
sort3 = '//*[@id="content"]/div[2]/ul/li[4]/a/span'
#학급경영
sort4 = '//*[@id="content"]/div[2]/ul/li[5]/a/span'
#ICT
sort5 = '//*[@id="content"]/div[2]/ul/li[6]/a/span'
#어학
sort6 = '//*[@id="content"]/div[2]/ul/li[7]/a/span'
#자기개발
sort7 = '//*[@id="content"]/div[2]/ul/li[8]/a/span'
```

# 23년 데이터 크롤링

```python
import time

ind = 1  # 현재 복사한 순서
year = driver.find_element(by='xpath',
                           value =f'//*[@id="content"]/div[3]/table/tbody/tr[{ind}]/td[6]' ).text.split('.')[0]
no = 1+1  # 1~10페이지 중 위치한 곳
page = 1  # page번호
list1 = []  # 결과물이 저장되는 리스트

#원하는 과정 설정(value 값 변경)
driver.find_element(by='xpath',value =sort7 ).click()
while 1:
    if int(year) > 2022:
        try:
            
            # 업체명, 주소
            driver.find_element(by='xpath',
                                value =f'//*[@id="content"]/div[3]/table/tbody/tr[{ind}]/td[3]/a' ).click()
            
            

            # 제목
            title = driver.find_element(by='xpath',
                                        value = f'//*[@id="bbs_wrap"]/div/dl[1]/dd').text
            # 연수명
            name = driver.find_element(by='xpath',
                                        value = f'//*[@id="bbs_wrap"]/div/dl[2]/dd/a').text
            # 만족도
            element = driver.find_element(by='xpath',value = f'//*[@id="bbs_wrap"]/div/dl[3]/dd[1]/span/span')
            star = element.get_attribute("class")
            # 분류
            sort = driver.find_element(by='xpath',
                                       value = f'//*[@id="bbs_wrap"]/div/dl[3]/dd[2]').text

            #등록일
            date = driver.find_element(by='xpath',
                                       value ='//*[@id="bbs_wrap"]/div/dl[4]/dd[2]' ).text


            # 본문내용
            content = driver.find_element(by='xpath',
                                          value = f'//*[@id="bbs_wrap"]/div/div[1]').text

            list1.append([title, name, star, sort, date, content]) #리스트 마지막에 데이터 추가
            ind += 1

            
            time.sleep(0.05)
            #다시 목록으로 돌아감
            driver.find_element(by='xpath',value ='//*[@id="btn-list"]' ).click()
            
            #연도 재설정
            driver.find_element(by='xpath',value = f'//*[@id="content"]/div[3]/table/tbody/tr[{ind}]/td[6]').is_displayed()
            year = driver.find_element(by='xpath',
                                       value =f'//*[@id="content"]/div[3]/table/tbody/tr[{ind}]/td[6]' ).text.split('.')[0]


        except:
            # 다음 리스트열로 넘어갈 수 있는지
            if no >=11: 
                ind = 1
                #driver.find_element(by='xpath',value = f'//*[@id="paging"]/ul/li[{no}]/a').click()# 2,3d으로 넘어감
                driver.find_element(by='xpath',value = f'//*[@id="paging"]/ul/li[12]/a').click() #11,21 로 넘어감

                page += 1
                no = (1+1)
                year = driver.find_element(by='xpath',
                                       value =f'//*[@id="content"]/div[3]/table/tbody/tr[{ind}]/td[6]' ).text.split('.')[0]

                continue


            # 다음 페이지로 이동 # 10페이지 단위마다 다음 페이지 버튼 누르기
            elif driver.find_element(by='xpath',value = f'//*[@id="paging"]/ul/li[{no+1}]/a').is_displayed():
                no += 1
                driver.find_element(by='xpath',value = f'//*[@id="paging"]/ul/li[{no}]/a').click()# 2,3d으로 넘어
                ind = 1
                page += 1 #10번대로 넘어감
                year = driver.find_element(by='xpath',
                                       value =f'//*[@id="content"]/div[3]/table/tbody/tr[{ind}]/td[6]' ).text.split('.')[0]
                continue



            # 더이상 이동할수 없을 경우 종료
            else :
                break
    else:
        break







```
