# 23년 데이터 크롤링

```python
import time

ind = 1  # 현재 복사한 순서
year = 2023
no = 1+1  # 1~10페이지 중 위치한 곳
page = 1  # page번호
list1 = []  # 결과물이 저장되는 리스트
main_window_handle = driver.current_window_handle
while 1:
    if int(year) > 2022:
        try:
          
            # 링크 클릭
            driver.find_element(by='xpath',
                                value =f'//*[@id="mainForm"]/table/tbody/tr[{ind}]/td[2]/a/strong' ).click()
            # 모든 창의 핸들 가져옴
            all_window_handles = driver.window_handles
            # 새로 열린 창으로 전환합니다.
            for window_handle in all_window_handles:
                if window_handle != main_window_handle:
                    driver.switch_to.window(window_handle)
                    break
            
            
            time.sleep(0.6)
            
            # 데이터 가져옴
           
            # 제목
            title = driver.find_element(by='xpath',
                                        value = f'//*[@id="mainForm"]/div[2]/table/tbody/tr[1]/td').text
            #연수명
            
            name = driver.find_element(by='xpath',
                                        value = f'//*[@id="mainForm"]/div[1]/span/a').text

            #만족도(별점)
            element = driver.find_element(by='xpath',value = f'//*[@id="mainForm"]/div[2]/table/tbody/tr[2]/td/span[1]/img')
            star = element.get_attribute("alt")
            
            #등록일
            date = driver.find_element(by='xpath',
                                        value = f'//*[@id="mainForm"]/div[2]/table/tbody/tr[2]/td/span[2]').text.split(' ')[-1]
            
            #내용
            content = driver.find_element(by='xpath',
                                      value = f'//*[@id="mainForm"]/div[2]/table/tbody/tr[3]/td').text
            
            list1.append([title, name, star, date, content]) #리스트 마지막에 데이터 추가
            ind += 1
            
            #확인버튼 클릭
            driver.find_element(by='xpath',
                                value = f'//*[@id="mainForm"]/div[2]/div/a' ).click()
            
           
            #원래 핸들로 돌아옴
            driver.switch_to.window(main_window_handle)
            time.sleep(0.1)
            
            #다음 강의평 연도
            driver.find_element(by='xpath',
                                value = f'//*[@id="mainForm"]/table/tbody/tr[{ind}]/td[3]').is_displayed()
            date_a=driver.find_element(by='xpath',
                    value = f'//*[@id="mainForm"]/table/tbody/tr[{ind}]/td[3]').text.split('\n')[-1]
            year=date_a.split('.')[0]

            
        except:
            # 다음 리스트열로 넘어갈 수 있는지
            if no >=11: 
                ind = 1
                #driver.find_element(by='xpath',value = f'//*[@id="paging"]/ul/li[{no}]/a').click()# 2,3d으로 넘어감
                driver.find_element(by='xpath',value = f'//*[@id="mainForm"]/div[3]/a[12]/img').click() #11,21 로 넘어감
             
                time.sleep(0.5)
                page += 1
                no = (1+1)
                date_a=driver.find_element(by='xpath',
                                           value = f'//*[@id="mainForm"]/table/tbody/tr[{ind}]/td[3]').text.split('\n')[-1]
                year=date_a.split('.')[0]

                continue


            # 다음 페이지로 이동 # 10페이지 단위마다 다음 페이지 버튼 누르기
            elif driver.find_element(by='xpath',value = f'//*[@id="mainForm"]/div[3]/a[{no+1}]').is_displayed():
                no += 1
                driver.find_element(by='xpath',value = f'//*[@id="mainForm"]/div[3]/a[{no}]').click()# 2,3d으로 넘어
                ind = 1
                page += 1 #10번대로 넘어감
                
                date_a=driver.find_element(by='xpath',
                                           value = f'//*[@id="mainForm"]/table/tbody/tr[{ind}]/td[3]').text.split('\n')[-1]
                year=date_a.split('.')[0]
                continue



            # 더이상 이동할수 없을 경우 종료
            else :
                break
    else:
        break







```
